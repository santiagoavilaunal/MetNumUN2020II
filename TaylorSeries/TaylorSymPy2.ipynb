```
from sympy import series, Symbol, latex
from IPython.display import Math
from sympy.functions import sin, cos, exp, ln
from sympy.plotting import plot
import matplotlib.pyplot as plt
```

```
x = Symbol('x')
```

```
#se crea una funcion que obtenga la serie de Tayor al ingresar una funcion utilizndo la libreria de sympy

def Taylor(funcion,x0,n):
  return funcion.series(x,x0,n).removeO()
```

```
#obtener la serie de Taylor de

#cos(x)
display(Math(r'cos(x) \approx '+latex(Taylor(cos(x),0,10))))

#sin(x)
display(Math(r'sin(x) \approx '+latex(Taylor(sin(x),0,10))))

#e^(x)
display(Math(r'e^x \approx '+latex(Taylor(exp(x),0,10))))

#1/(1-x)
display(Math(r'\frac{1}{1-x} \approx '+latex(Taylor(1/(1-x),0,10))))

#ln(1+x)
display(Math(r'ln(1+x) \approx '+latex(Taylor(ln(x+1),0,10))))

#ln(x)
display(Math(r'ln(x) \approx '+latex(Taylor(ln(x),1,10))))
```

```
def plot_Taylo(funcion,x0,n):
  p = plot(funcion,legend=True, show=False)
  p.append(plot(Taylor(funcion,x0,2+n),show=False)[0])
  p.append(plot(Taylor(funcion,x0,3+n),show=False)[0])
  p.append(plot(Taylor(funcion,x0,4+n),show=False)[0])
  p.append(plot(Taylor(funcion,x0,5+n),show=False)[0])

  p[0].line_color = 'blue'
  p[1].line_color = 'green'
  p[2].line_color = 'firebrick'
  p[3].line_color = 'black'
  p[4].line_color = 'orange'
  p.xlim=(-3,3)
  p.ylim=(-5,5)
  p.title="Funcion {0} y Taylo".format(funcion)
  p.show()
```

```
#cos(x)
display(Math('cos(x)'))
plot_Taylo(cos(x),0,1)

#sin(x)
display(Math('sin(x)'))
plot_Taylo(sin(x),0,0)

#1/(1-x)
display(Math(r'\frac{1}{1-x}'))
plot_Taylo((1/(1-x)),0,0)

#ln(1+x)
display(Math(r'ln(1+x)'))
plot_Taylo(ln(x+1),1,0)

#ln(1+x)
display(Math(r'ln(x)'))
plot_Taylo(ln(x),1,0)

#e^x
display(Math(r'e^x'))
plot_Taylo(exp(x),0,0)
```
